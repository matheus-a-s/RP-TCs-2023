```python
import pandas as pd
import numpy as np

data_path = "./dataset/ionosphere.data"

data = pd.read_csv(data_path, delimiter=",", header=None)
data
```

```python
data_values = data.iloc[:,:-1]
data_values
```

```python
print("Número de atributos", len(data_values.columns))
print("Número de medições", len(data_values))
print("Médias", data_values.mean())
```

## Questão 1

```python
def cov_matrix(data):
    num_attrs = len(data.columns)
    num_msmts = len(data)
    mean = data.mean()

    cx = np.zeros((num_attrs, num_attrs))

    for i in range(num_attrs):
        for j in range(num_attrs):
            aux = np.expand_dims(np.asarray(data.iloc[:, i]) - mean[i], axis=1)
            aux_t = np.expand_dims(np.asarray(data.iloc[:, j]) - mean[j], axis=1)
            cx[i][j] = (aux_t * aux).sum() / num_msmts
            
    return cx

def fast_cov_matrix(data):
    num_attrs = len(data.columns)
    num_msmts = len(data)
    mean = data.mean()

    aux = np.asarray(data - mean)
    cx = (aux.T @ aux) / num_msmts
    return cx

def alternative_cov_matrix(data):
    num_attrs = len(data.columns)
    num_msmts = len(data)
    mean = np.expand_dims(data.mean(), axis=1)

    rx = np.zeros((num_attrs, num_attrs))

    for i in range(num_attrs):
        for j in range(num_attrs):
            aux = np.expand_dims(np.asarray(data.iloc[:, i]), axis=1)
            aux_t = np.expand_dims(np.asarray(data.iloc[:, j]), axis=1)
            rx[i][j] = (aux_t * aux).sum() 
            
    return (rx / num_msmts) - (mean * mean.T)

def fast_alternative_cov_matrix(data):
    num_attrs = len(data.columns)
    num_msmts = len(data)
    mean = np.expand_dims(data.mean(), axis=1)

    aux = np.asarray(data)
    rx = (aux.T @ aux) / num_msmts
    return rx - (mean * mean.T)
```

```python
cx = cov_matrix(data_values)
fast_cx = fast_cov_matrix(data_values)
alt_cx = alternative_cov_matrix(data_values)
fast_alt_cx = fast_alternative_cov_matrix(data_values)

ref_cx = np.cov(data_values, rowvar=False)

print("Matriz de covariância implementada \n", cx)
print("Matriz de covariância implementada rápida \n", fast_cx)

print("Matriz de covariância alternativa implementada \n", alt_cx)
print("Matriz de covariância alternativa implementada rápida \n", fast_alt_cx)

print("Matriz de covariância de referência \n", ref_cx)

print("Norma (Frobenius) da matriz de diferenças: ", np.linalg.norm(cx - ref_cx, 'fro')) # Frobenius 
print("Norma (Frobenius) da matriz de diferenças (rápida): ", np.linalg.norm(fast_cx - ref_cx, 'fro')) # Frobenius 

print("Norma (Frobenius) da matriz de diferenças (alternativa): ", np.linalg.norm(alt_cx - ref_cx, 'fro')) # Frobenius 
print("Norma (Frobenius) da matriz de diferenças (rápida alternativa): ", np.linalg.norm(fast_alt_cx - ref_cx, 'fro')) # Frobenius 
```

## Questão 4

```python
cond_number_global_cx = np.linalg.cond(fast_cx)
cond_number_good_cx = np.linalg.cond(g_cx)
cond_number_bad_cx = np.linalg.cond(b_cx)

rank_global_cx = np.linalg.matrix_rank(fast_cx)
rank_good_cx = np.linalg.matrix_rank(g_cx)
rank_bad_cx = np.linalg.matrix_rank(b_cx)

print("Num de condicionamento CX Global:", cond_number_global_cx)
print("Num de condicionamento CX Good:", cond_number_good_cx)
print("Num de condicionamento CX Bad:", cond_number_bad_cx)

print("Posto CX Global:", rank_global_cx)
print("Posto CX Good:", rank_good_cx)
print("Posto CX Bad:", rank_bad_cx)
```
